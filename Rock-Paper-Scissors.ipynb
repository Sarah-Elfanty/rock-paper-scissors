```python
# Write your code here
import random

score = 0
scores = {}
exit = 0
moves = []
options = {"!exit": 0,
           "!rating":0}
all_moves = {'air':['scissors', 'snake', 'human', 'tree', 'wolf', 'sponge','paper'],
            'water':['snake', 'human','tree', 'wolf', 'sponge', 'paper', 'air'],
            'dragon':['human','tree', 'wolf', 'sponge', 'paper', 'air', 'water'],
            'devil':['tree', 'wolf', 'sponge', 'paper', 'air', 'water', 'dragon'],
            'lightning':['wolf', 'sponge', 'paper', 'air', 'water', 'dragon', 'devil'],
            'gun':['sponge', 'paper', 'air', 'water', 'dragon', 'devil','lightning'],
            'rock':['paper', 'air', 'water', 'dragon', 'devil','lightning','gun'],
            'fire':['air', 'water', 'dragon', 'devil','lightning','gun', 'rock'],
            'scissors':['water', 'dragon', 'devil','lightning','gun', 'rock', 'fire'],
            'snake':['dragon', 'devil','lightning','gun', 'rock', 'fire', 'scissors'],
            'human':['devil','lightning','gun', 'rock', 'fire', 'scissors', 'snake'],
            'tree':['lightning', 'gun', 'rock', 'fire', 'scissors', 'snake', 'human'],
            'wolf':['gun', 'rock', 'fire', 'scissors', 'snake', 'human', 'tree'],
            'sponge':['rock', 'fire', 'scissors', 'snake', 'human', 'tree', 'wolf'],
            'paper':['fire', 'scissors', 'snake', 'human', 'tree', 'wolf', 'sponge']}

user_in = input("Enter your name: ")

print("Hello, {}".format(user_in))

scores_file = open('test.txt', 'r')

for line in scores_file:
    person = line.replace('\n', '').split(' ')
    #print(person)
    scores[person[0]] = person[1]
scores_file.close()

if user_in in scores.keys():
    score = int(scores[user_in])


userin = input()
if userin == '':
    options['paper'] = 'scissors'
    options['scissors'] = 'rock'
    options['rock'] = 'paper'
    moves = ['paper', 'rock', 'scissors']
else:
    
    for option in userin.split(','):
        options[option] = all_moves[option]
        moves.append(option)
    #print(options)
print("Okay, let's start")


    
while not exit:
    comp_move = random.choice(moves)
    user_in = input()
    if user_in not in options.keys():
        print("Invalid input")
    else:
        if user_in == '!exit':
            exit = 1
            break
        elif user_in == '!rating':
            print("Your rating: {}".format(score))
        elif user_in == comp_move:
            print("There is a draw ({})".format(user_in))
            score += 50
        elif comp_move in all_moves[user_in]:
            print("Sorry, but computer chose", comp_move)
        else:
            print("Well done. Computer chose",comp_move,"and failed")
            score += 100
print("Bye!")
```
